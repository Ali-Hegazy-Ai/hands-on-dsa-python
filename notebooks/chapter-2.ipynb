```python
import pandas as pd

df = pd.read_csv("asn-ipv4.csv", header=None)

first_column = df.iloc[:, 0]

Ip_Regions = {}

for ip in first_column:
    region = ip.split(".")[:2]
    region = ".".join(region)
    if region not in Ip_Regions:
        Ip_Regions[region] = 1
    else:
        Ip_Regions[region] += 1

# Convert the dictionary to a DataFrame and display as a table
region_df = pd.DataFrame(list(Ip_Regions.items()), columns=["Region", "Count"])
display(region_df)
```
